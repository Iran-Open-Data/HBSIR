# Loading data

## Load Standard Data

```python
data_engine.load_table('food', 1400)
```

## Load Original Data

```python
data_engine.load_table('food', 1400, standard=False)
```

## Add Year to Data

```python
data_engine.load_table('food', 1400, add_year=True)
```

**Note**

- If only one year loaded, the default value of add_year is false, otherwise it's true.

## Load Data for Range of Years

```python
data_engine.load_table('food', 1396, 1400)
```

# Adding Attributes and Translations

## Check Available Attributes

```python
data_engine._Attributes
```

## Add Household Attributes to Table

```python
year = 1400

df = data_engine.load_table('food', year)

data_engine.add_attribute(df, year, ["Urban-Rural", "Province"])
```

## Add Commodity Code Classification to Table

```python
data_engine.add_classification(df, classification="ORIGINAL_1383", level=2, year=year)
```
